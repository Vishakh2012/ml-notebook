```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
#importing required modules
```

```python
data = pd.read_csv('Housing.xls')
data
#importing required data 
```

```python
x = data['area']
y = data['price']
new_data = [x,y]
new_data = pd.concat(new_data,axis=1)
new_data

```

```python
X = np.array(x).reshape(1,-1)
Y = np.array(y).reshape(-1,1)
plt.figure(figsize = (20,20))
plt.scatter(X,Y)
plt.title('price vs size of a house')
plt.xlabel('input : size')
plt.ylabel('output : price')




```

```python
class linear_coeff():
    def __init__(self, feature, label):
        self.feature = X
        self.label = y
    global x_mean, y_mean, n
    n = x.shape[0]  # returns ther number of features in the dataset
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    xar_mean = np.array(x_mean)
    yar_mean = np.array(y_mean)
    num = 0
    dem = 0

    def weight_of_line(self):
        num = np.matmul((np.array(x) - np.array(x_mean)),(np.array(y)-np.array(y_mean)))
        dem = np.matmul((np.array(x) - np.array(x_mean)),(np.array(x) - np.array(x_mean)))
        weight = num/dem
        return weight

    def bias_of_line(self):

        num = np.matmul((np.array(x) - np.array(x_mean)),(np.array(y)-np.array(y_mean)))
        dem = np.matmul((np.array(x) - np.array(x_mean)),(np.array(x) - np.array(x_mean)))

        weight = num/dem

        bias = y_mean-np.dot(weight,x_mean)
        return bias

    def required_y_hat(self):
        W = self.weight_of_line()
        B = self.bias_of_line()
        y_hat = np.dot(W,x) + B
        return y_hat
```

```python

X = np.array(x).reshape(1,-1)
Y = np.array(y).reshape(-1,1)
plt.figure(figsize = (20,20))
plt.scatter(X,Y)
data = linear_coeff(x,y)
y_hat = data.required_y_hat()
plt.plot(x,y_hat, color = 'green', linewidth = '4')
plt.title('price vs size of a house')
plt.xlabel('input : size')
plt.ylabel('output : price')
        
```
