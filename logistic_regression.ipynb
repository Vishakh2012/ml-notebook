```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt

```

```python
input_arr = np.array([2,4,6])
output_arr = np.exp(input_arr)
print(f"input array is {input_arr}")
print(f"the output array is {output_arr}")
```

```python
def sigmoid(z):
    """sigmoid is a function that computes the sigmoid function
    input: ndarray for which sigmoid is calculated
    output: ndarray"""
    k = pow(1 + np.exp(-z),-1)
    return k
```

```python
z_temp = np.arange(-10,11)#an array of equally spaced numbers between -10 an 10
print(z_temp)
y = sigmoid(z_temp)
np.set_printoptions(precision = 3)#sets the number of digits after decimal - precision
print("[input z],[output y]")
print(np.c_[z_temp, y])
```

```python
plt.figure(figsize = (3,3))
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.plot(z_temp,y)
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z_temp, y, c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Load the data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [3, 4, 5, 6, 7]

# Create the figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Plot the data
ax.scatter3D(x, y, z, c=z, cmap='Reds')

# Add labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:#o prints circles while ^ prints triangles
    
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
